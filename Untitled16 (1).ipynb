**Data** **Preprocessing**

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

IMG_SIZE = 224
BATCH_SIZE = 32

# Data augmentation + normalization
train_datagen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2  # 20% for validation
)

# Training generator
train_generator = train_datagen.flow_from_directory(
    'tb_xray/TB_Chest_Radiography_Database',
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='binary',
    subset='training'
)

# Validation generator
val_generator = train_datagen.flow_from_directory(
    'tb_xray/TB_Chest_Radiography_Database',
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='binary',
    subset='validation'
)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
model = Sequential([Conv2D(32,(3,3), activation='relu',
input_shape=(224,224,3)),
MaxPooling2D(2,2),

Conv2D(64,(3,3),activation='relu'),
MaxPooling2D(2,2),

Conv2D(128,(3,3),activation='relu'),
MaxPooling2D(2,2),

Flatten(),
Dense(128,activation='relu'),
Dropout(0.5),
Dense(1,activation='sigmoid')
                    ])
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.summary()

```

**Training**

```python
history = model.fit(
    train_generator,
    steps_per_epoch = 100// BATCH_SIZE,
    epochs = 5,
    validation_data = val_generator,
    validation_steps = 50 // BATCH_SIZE
)
```

**Feature Extraction of ML**

```python
import tensorflow as tf
from tensorflow.keras.applications import VGG16
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.vgg16 import preprocess_input, decode_predictions
import numpy as np
base_model = VGG16(weights='imagenet',
include_top= False, input_shape=(224,224,3))
base_model.trainable = False

feature_extractor = tf.keras.Sequential([
base_model,
tf.keras.layers.Flatten()
])

data_dir = 'tb_xray/TB_Chest_Radiography_Database'
datagen_ml = ImageDataGenerator(preprocessing_function = preprocess_input)

ml_gen = datagen_ml.flow_from_directory(
     data_dir,
     target_size = (224,224),
     batch_size = 32,
     class_mode = 'binary',
     shuffle = False
 )

X_features = feature_extractor.predict(ml_gen)
y_labels = ml_gen.classes

print("Features extracted:", X_features.shape)
print("Labels:", y_labels.shape)
```

**Train ML Models**

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
X_train, X_test, y_train, y_test = train_test_split(X_features, y_labels, test_size=0.2, random_state=42)

results = {}

#Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X_train, y_train)
results["Random Forest"] = accuracy_score(y_test, rf.predict(X_test) )

#SVM
svm = SVC(kernel='linear', random_state=42)
svm.fit(X_train, y_train)
results["SVM"] = accuracy_score(y_test,svm.predict(X_test))

#Logistic Regression
lr = LogisticRegression(random_state=42, max_iter = 200)
lr.fit(X_train, y_train)
results["Logistic Regression"] = accuracy_score(y_test, lr.predict(X_test))

#KNN
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
results["KNN"] = accuracy_score(y_test, knn.predict(X_test))

print("ML Model Accuracies:")
for model, acc in results.items():
  print(f"{model}:{acc:.4f}")
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

# Define CNN
cnn_model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(224,224,3)),
    MaxPooling2D(2,2),
    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid')
])

cnn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train quickly
history = cnn_model.fit(train_generator, validation_data=val_generator, epochs=3)

# Evaluate
loss, accuracy = cnn_model.evaluate(val_generator)
cnn_acc = accuracy
```

**Evaluation Metrics: CNN**

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
import seaborn as sns
import matplotlib.pyplot as plt

# Get CNN predictions
y_true_cnn = val_generator.classes
y_pred_cnn_prob = cnn_model.predict(val_generator)  #  use cnn_model here
y_pred_cnn = (y_pred_cnn_prob > 0.5).astype("int32")

# Confusion Matrix
cm = confusion_matrix(y_true_cnn, y_pred_cnn)
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d', cmap="Blues", xticklabels=["Normal","TB"], yticklabels=["Normal","TB"])
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("CNN Confusion Matrix")
plt.show()

# Classification Report
print("CNN Classification Report:\n")
print(classification_report(y_true_cnn, y_pred_cnn, target_names=["Normal","TB"]))

# ROC Curve
fpr, tpr, _ = roc_curve(y_true_cnn, y_pred_cnn_prob)
roc_auc = auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, label=f"CNN (AUC = {roc_auc:.2f})")
plt.plot([0,1], [0,1], "r--")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("CNN ROC Curve")
plt.legend()
plt.show()
```

**Evaluation Metrics: ML**

```python
from sklearn.metrics import roc_auc_score

for model_name, model in [("Random Forest", rf), ("SVM", svm), ("Logistic Regression", lr), ("KNN", knn)]:
    y_pred_ml = model.predict(X_test)
    y_pred_ml_prob = model.predict_proba(X_test)[:,1] if hasattr(model, "predict_proba") else None

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred_ml)
    plt.figure(figsize=(5,4))
    sns.heatmap(cm, annot=True, fmt='d', cmap="Greens", xticklabels=["Normal","TB"], yticklabels=["Normal","TB"])
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.title(f"{model_name} Confusion Matrix")
    plt.show()

    # Classification Report
    print(f"{model_name} Classification Report:\n")
    print(classification_report(y_test, y_pred_ml, target_names=["Normal","TB"]))

    # ROC Curve
    if y_pred_ml_prob is not None:
        fpr, tpr, _ = roc_curve(y_test, y_pred_ml_prob)
        roc_auc = roc_auc_score(y_test, y_pred_ml_prob)

        plt.figure()
        plt.plot(fpr, tpr, label=f"{model_name} (AUC = {roc_auc:.2f})")
        plt.plot([0,1], [0,1], "r--")
        plt.xlabel("False Positive Rate")
        plt.ylabel("True Positive Rate")
        plt.title(f"{model_name} ROC Curve")
        plt.legend()
        plt.show()
```
